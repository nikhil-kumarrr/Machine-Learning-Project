Encoding 

```python
en_data = df[["sex","smoker","region"]]
```

```python
clean_obj = pd.get_dummies(en_data,dtype="int",drop_first="True")
```

```python
clean_obj
```

```python
numerical
```

```python
clean_obj["age"] = numerical["age"]
```

```python
clean_obj["bmi"] = numerical["bmi"]
```

```python
clean_obj["children"] = numerical["children"]
```

```python
clean_obj["charges"] = numerical["charges"]
```

```python
clean_obj
```

Divide the data in training and testing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = clean_obj.drop(['charges'], axis='columns')
y = clean_obj[['charges']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 42,test_size=50,shuffle=True)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X.columns
```

```python
X_train= pd.DataFrame(X_train_scaled,columns=X.columns)
X_test= pd.DataFrame(X_test_scaled,columns=X.columns)
```

Build Model

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
krr= KNeighborsRegressor(n_neighbors=3)
```

```python
krr.fit(X_train,y_train)
```

```python
krr.score(X_train,y_train)*100
```

```python
krr.score(X_test,y_test)*100
```
